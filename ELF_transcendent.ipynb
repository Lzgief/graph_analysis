```python
import lief # Парсинг ELF
import os   # Парсинг папок
```

```python
# Файл с самым большим кол-вом импортов
for item in os.scandir(path='/bin'):
    if lief.is_elf('/bin/'+item.name):
        elf = lief.parse('/bin/'+item.name)
        print(item.name)
        for entry in elf.dynamic_entries:          
            if str(entry)[:6] == "NEEDED":
                print(entry)
        print('')
        
```
