```python
import pefile
import os
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
# Модуль Сканирование PE файла и получение импортируемых имен dll
def get_imports(path_to_file):
    pe = pefile.PE(path_to_file)
    dll = []
    for entry in pe.DIRECTORY_ENTRY_IMPORT:
        dll.append(entry.dll.decode('utf-8').lower())
    return set(dll)
```

```python
# Создать фрейм
# Столбцы - dll
# Строки - пути к файлам
# Пересечение - наличие связи

# Модуль сканирования системы
G = nx.Graph()
tree = []
for i in os.walk('cmd'):
    tree.append(i)

system32 = []
for i in os.walk('c:\\windows\\system32'):
    system32.append(i)

# Модуль определения типов файлов
for address, dirs, files in tree:
    for file in files:
        #print(address+'\\'+file)
        path = address+'\\'+file
        file_name, file_ext = os.path.splitext(path)
        
        # Проверить если файл PE расширения
        if file_ext in ('.exe','.dll'):
            imps = get_imports(path)
            for imp in imps:
                if imp in files:
                    G.add_edge(path, address+'\\'+imp)
                if imp in system32[0][2]:
                    G.add_edge(path, system32[0][0]+'\\'+imp)
                
G.nodes()
```

```python
EXE = nx.Graph()
for node in G.nodes:
    path, ext = os.path.splitext(node)
    if ext == '.exe':
        EXE.add_node(node, mtime = os.path.getmtime(node))
EXE.nodes(data=True)
        
```

```python
DLL = nx.Graph()
for node in G.nodes:
    path, ext = os.path.splitext(node)
    if ext == '.dll':
        DLL.add_node(node, mtime = os.path.getmtime(node), uid=os.stat(node).st_uid)
DLL.nodes(data=True)
        
```

```python
pos_dot = graphviz_layout(G, prog='dot')
```

```python
fig = plt.figure(figsize=(16, 8))
exe = nx.draw_networkx_nodes(
    EXE, pos_dot, node_size=2, node_shape='o', node_color='grey')
dll = nx.draw_networkx_nodes(
    DLL, pos_dot, node_size=2, node_shape='d',
    node_color=[color[1]['mtime'] for color in DLL.nodes(data=True)],
    cmap = 'hot_r')
edges = nx.draw_networkx_edges(G, pos_dot, edge_color='wheat', width=0.02)
labels = nx.draw_networkx_labels(DLL, pos_dot, font_size=1)
for _,t in labels.items():
    t.set_rotation('vertical')
plt.axis('off')
fig.savefig('fig_neaot.png', dpi=700)
```
