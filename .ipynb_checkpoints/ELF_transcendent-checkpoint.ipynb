```python
G = nx.DiGraph()
start_path = pathlib.Path("/bin/zettlr")
for lib in libs:
    if lib[8:9] != "/" : # Отсеивание либ не из /usr/lib/
        continue
    #print(lib)    
    G.add_edge(str(start_path), str(lib))
    path = pathlib.Path(lib)
    #print(path)
    #print('')
    deps = lddwrap.list_dependencies(path=path)
    for dep in deps:
        if (str(dep.path)[8:9] == "6") or (dep.path == None): # Отсеивание либ не из /usr/lib/
            continue
        #print(dep.path)
        if dep.path == None:
            G.add_edge(str(lib), str(dep.soname))
        else:
            G.add_edge(str(lib), str(dep.path))
```

```python
def output_PageRank(iteration, graph, damping_factor):
    PageRank(graph, damping_factor, iteration)
    pagerank_list = graph.get_pagerank_list()
    print('PageRank:')
    print(pagerank_list)
    return pagerank_list
```

# А теперь будем считать для каждой программы!!!!! ВУХААХАХА

```python
def get_libs(bin_path):
    libs = []
    deps = lddwrap.list_dependencies(bin_path)
    for dep in deps:
        if dep.path != None:
            libs.append(str(dep.path))
    return libs
```

```python
def mkG(ELF):
    G = nx.DiGraph()
    for lib in ELF[1]:
        if lib[8:9] != "/":
            continue
        G.add_edge(ELF[0], lib)
        path = pathlib.Path(lib)
        deps = lddwrap.list_dependencies(path=path)
        for dep in deps:
            if (str(dep.path)[8:9] == "6") or (dep.path == None): 
                continue
            if dep.path == None:
                G.add_edge(lib, str(dep.soname))
            else:
                G.add_edge(lib, str(dep.path))
    return G
```

```python
def mkLG(ELF):
    LG = Graph()
    for lib in ELF[1]:
        if lib[8:9] != "/":
            continue
        LG.add_edge(ELF[0], lib)
        path = pathlib.Path(lib)
        deps = lddwrap.list_dependencies(path=path)
        for dep in deps:
            if (str(dep.path)[8:9] == "6") or (dep.path == None): 
                continue
            if dep.path == None:
                LG.add_edge(lib, str(dep.soname))
            else:
                LG.add_edge(lib, str(dep.path))
    return LG
```

```python
def draw_plots(Gs, clrPR, clrAuth, clrHubs):
    cmap = "bwr"
    i = 0
    for G in Gs:      
        plt.figure(1)
        nx.draw_networkx(G, node_color=clrPR[i], with_labels = False, cmap = cmap)
        plt.savefig("out/PR"+str(i)+".png")
        plt.figure(2)
        nx.draw_networkx(G, node_color=clrAuth[i], with_labels = False, cmap = cmap)
        plt.savefig("out/Auth"+str(i)+".png")
        plt.figure(3)
        nx.draw_networkx(G, node_color=clrHubs[i], with_labels = False, cmap = cmap)
        plt.savefig("out/Hub"+str(i)+".png")
        i = i + 1
```

```python
Elves = {}
for item in os.scandir(path='/bin'):
    if lief.is_elf('/bin/'+item.name):
        path = pathlib.Path("/bin/"+item.name)
        try:
            Elves.update({str(path):get_libs(path)})
        except: pass 
```

```python
Gs = []
LGs = []
for ELF in Elves.items():
    try:
        Gs.append(mkG(ELF))
        LGs.append(mkLG(ELF))
    except: pass
```

```python
clrPR = []
clrHubs = []
clrAuth = []
iters = 100
dmp_fac = 0.15
for LG in LGs:
    clrPR.append(output_PageRank(iters, LG, dmp_fac))
    clrAuth.append(output_HITS(iters,LG)[0])
    clrHubs.append(output_HITS(iters,LG)[1])
```

```python
draw_plots(Gs, clrPR, clrAuth, clrHubs)
```
